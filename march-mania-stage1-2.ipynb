```python
#import all required packages
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from pandas import *
from numpy import nan as NA
import statsmodels.api as sm
import statsmodels.formula.api as smf
import seaborn as sns
import imblearn
from sklearn.preprocessing import RobustScaler
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn import metrics
```

```python
#import all files from data section1
MTeams=pd.read_csv('MTeams.csv')
WTeams=pd.read_csv('WTeams.csv')
MSeasons=pd.read_csv('MSeasons.csv')
WSeasons=pd.read_csv('WSeasons.csv')
MSeeds=pd.read_csv('MNCAATourneySeeds.csv')
WSeeds=pd.read_csv('WNCAATourneySeeds.csv')
MResults=pd.read_csv('MRegularSeasonCompactResults.csv')
WResults=pd.read_csv('WRegularSeasonCompactResults.csv')
MTourney=pd.read_csv('MNCAATourneyCompactResults.csv')
WTourney=pd.read_csv('WNCAATourneyCompactResults.csv')

print(MSeasons.head()) #the season dataset does not have any Team ID, and only has the date game started and eash region --> could not be easily joined with other data sets

#join the seasons information and seeds datasets:
male_new_data=pd.merge(MTeams,MSeeds,on='TeamID')
print(male_new_data.head())
#concatenate the regular season and tournament results:
male_new_results=pd.concat([MResults,MTourney])
#note for Tournament games all men's games were on neutral site
#so now the results are stacked together
print(male_new_results.head())


#join the female teams and seeds datasets:
female_new_data=pd.merge(WTeams,WSeeds,on='TeamID')
#print(male_new_data)
#concatenate the regular season and tournament results:
female_new_results=pd.concat([WResults,WTourney])

#for male_new_results, change Season, WTeamID, LTeamID to categorical variable
male_new_results['Season']=male_new_results['Season'].astype(str)
male_new_results['WTeamID']=male_new_results['WTeamID'].astype(str)
male_new_results['LTeamID']=male_new_results['LTeamID'].astype(str)

#for female_new_results, change Season, WTeamID, LTeamID to categorical variable
female_new_results['Season']=female_new_results['Season'].astype(str)
female_new_results['WTeamID']=female_new_results['WTeamID'].astype(str)
female_new_results['LTeamID']=female_new_results['LTeamID'].astype(str)

```

```python
#exploratory data analysis
#checking for missing values:
null1=pd.DataFrame({'Count': male_new_data.isnull().sum(), 'Percentage': male_new_data.isnull().sum()/len(male_new_data)})
null2=pd.DataFrame({'Count': male_new_results.isnull().sum(), 'Percentage': male_new_results.isnull().sum()/len(male_new_results)})
null3=pd.DataFrame({'Count': female_new_data.isnull().sum(), 'Percentage': female_new_data.isnull().sum()/len(female_new_data)})
null4=pd.DataFrame({'Count': female_new_results.isnull().sum(), 'Percentage': female_new_results.isnull().sum()/len(female_new_results)})
print(null1) #no missing values
print(null2) #no missing values
print(null3) #no missing values
print(null4) #no missing values

#create statistical summary
male_new_results.describe()

#create a histogram that counts the number of each Team appeared as winning team for male:
male_wteam_count=male_new_results.groupby('WTeamID').size()
print(male_wteam_count[male_wteam_count==male_wteam_count.max()]) #that the team that has most times of winning is 1181
plt.hist(male_wteam_count,color='skyblue')

#create a histogram that counts the number of each Team appeared as winning team for female:
female_wteam_count=female_new_results.groupby('WTeamID').size()
print(female_wteam_count[female_wteam_count==female_wteam_count.max()]) #that the team that has most times of winning is 3163
plt.hist(female_wteam_count,color='skyblue')

#now use the Season data sets as a map, translate all the seed to the region they were in
```
